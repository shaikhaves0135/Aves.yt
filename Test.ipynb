```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("/content/Breast Cancer Data Set.csv")
```

```python
df
```

```python
df.info()
```

```python
df.drop(["Unnamed: 32","id"],axis = 1,inplace = True)
```

```python
df.describe()
```

```python
df["diagnosis"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
Lb = LabelEncoder()
df["diagnosis"] = Lb.fit_transform(df["diagnosis"])
```

```python
x = df.drop("diagnosis",axis = 1)
y = df["diagnosis"]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm = MinMaxScaler()
cols = ["radius_mean","texture_mean","perimeter_mean","area_mean","radius_worst","texture_worst","perimeter_worst","area_worst"]
x[cols] = mm.fit_transform(x[cols])
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,random_state=1,test_size = 0.25,stratify = y)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
ann = Sequential()

ann.add(Dense(units = 30, activation = "relu"))
ann.add(Dense(units = 15, activation = "relu"))
ann.add(Dense(units = 1, activation = "sigmoid"))

ann.compile(loss = "binary_crossentropy",optimizer = "adam")

ann.fit(xtrain,ytrain,validation_data = (xtest,ytest),epochs = 100,batch_size = 20)
```

```python
lossdf = pd.DataFrame(ann.history.history)
lossdf.plot()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
earlystopping = EarlyStopping(patience = 10,mode = "min",monitor = "val_loss",verbose = 1)
```

```python
ann = Sequential()

ann.add(Dense(units = 30, activation = "relu"))
ann.add(Dense(units = 15, activation = "relu"))
ann.add(Dense(units = 1, activation = "sigmoid"))

ann.compile(loss = "binary_crossentropy",optimizer = "adam")
```

```python
ann.fit(xtrain,ytrain,validation_data = (xtest,ytest),epochs = 100,batch_size = 20,callbacks=[earlystopping])
```

```python
lossdf = pd.DataFrame(ann.history.history)
lossdf.plot()
```

```python
from tensorflow.keras.layers import Dropout
ann = Sequential()

ann.add(Dense(units = 30, activation = "relu"))
ann.add(Dropout(rate = 0.5))
ann.add(Dense(units = 15, activation = "relu"))
ann.add(Dropout(rate = 0.5))
ann.add(Dense(units = 1, activation = "sigmoid"))

ann.compile(loss = "binary_crossentropy",optimizer = "adam")
ann.fit(xtrain,ytrain,validation_data = (xtest,ytest),epochs = 100,batch_size = 20)
```

```python
lossdf = pd.DataFrame(ann.history.history)
lossdf.plot()
```
